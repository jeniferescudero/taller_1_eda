## 2. Entendimiento inicial de datos:

Se determinan las dimensiones de la base de datos, es decir, se realiza el conteo de registros (Filas) y variables (Columnas), así como el tipo de datos, tipo de variables de cada una de ellas, el porcentaje de datos nulos por variable.

```python
def caracteristicas_dtf(name):
    print('Las dimensiones del conjunto de datos son:', name.shape, 'correspondiente a',
          name.shape[0], 'registros y', name.shape[1] , 'columnas (variables)')
    print(end = "\n")
    print('El listado de variables para el conjunto de datos es: ',"\n", name.keys())
    print(end = "\n")
    print('Los tipos de variable leido del conjunto de datos son:',"\n")
    print(name.info())
    print(end = "\n")
    print('La cantidad de valores únicos y porcentaje de nulidad por variable son:',"\n")
    for col in name.columns:
     print(col, ' - tiene', name[col].nunique(), 'valores únicos' , 'y', f"{round(np.mean(name[col].isnull())*100)}%", 'de datos nulos')
    print(end = "\n")
    print(f'El conjunto de datos tiene {name.duplicated().sum()} registros(filas) duplicados.')
    #print('El porcentaje de datos nulos para cada variable del conjunto de datos es:',"\n")
    #for col in name.columns:
     #   pct_missing = np.mean(name[col].isnull())
      #  print(f"{col} - {round(pct_missing*100)}%")
caracteristicas_dtf(df)
```

## 3. Estrategia de análisis

En el análisis del conjunto de datos para establecer  algunas opciones de inversion inmobiliaria se tendra en cuenta la ubicación, el tipo de habitaciones, el precio por noche y la ocupación media de las propiedades en Praga, esto con la finalidad de identificar las zonas y tipos de habitaciones más atractivas para los inversores, utilizamos los siguientes criterios:

- Precio por noche: Las zonas y tipos de habitaciones con un precio por noche más alto son más atractivas para los inversores, ya que generan mayores ingresos.

- Ocupación media: Las zonas y tipos de habitaciones con una ocupación media más alta son también más atractivas para los inversores, ya que ofrecen una mayor probabilidad de éxito.

para esto podemos apoyarnos en metricas como

- Ingresos potenciales: Esta métrica se calcula multiplicando el precio por noche por la ocupación media.
Para calcular los ingresos potenciales de una propiedad, podemos utilizar la siguiente fórmula:
Ingresos potenciales = Precio por noche * Ocupación media

Esta métrica proporciona al inversor una visión más completa del potencial de rendimiento de una inversión en Airbnb en Praga.
